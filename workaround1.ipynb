```python
def summed(arr):
    sums = []
    for i in range(len(arr)):
        sums.append(np.sum(arr[i]))
    return sums
```

```python
def diff_list(l1,l2):
    l3 = []
    for i in range(len(l1)):
        l3.append(l1[i]-l2[i])
    return l3
```
